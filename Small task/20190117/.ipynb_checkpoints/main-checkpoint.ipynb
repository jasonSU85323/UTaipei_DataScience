```python
import re
import time
import io
import os
import json
import pandas as pd
from optparse import OptionParser
from pandas import DataFrame, read_csv
```

```python
# 先匯入資料
allData = pd.read_csv(r'output.csv')
print(type(allData))
allData.head()
```

```python
allData.groupby(['class', 'name'])['content'].apply(' '.join).reset_index().head()
```

```python
mergeData = allData.groupby(['class', 'name'])['content'].apply(' '.join).reset_index()
mergeData.head()
```

```python
mergeData['class'][mergeData['class'] == "中式"] = 'Chinese'
mergeData['class'][mergeData['class'] == "西式"] = 'Western'
mergeData.head()
```

```python
print(type(mergeData['content']))
corpus = mergeData['content'].values.tolist()
print(type(corpus))
# print(corpus)
```

```python
# TF-IDF
# coding:utf-8  
from sklearn.feature_extraction.text import CountVectorizer  # 詞頻向量化
vectorizer = CountVectorizer()  
X = vectorizer.fit_transform(corpus)  # 計算各個詞語出現的次數
word = vectorizer.get_feature_names()  # 獲取詞袋中所有文本的關鍵字
print(X)

from sklearn.feature_extraction.text import TfidfTransformer  # TF-IDF處理
transformer = TfidfTransformer()  
tfidf = transformer.fit_transform(X)
print(tfidf)
```

```python
len(corpus)
print(len(corpus))
features = tfidf.toarray()
labels = mergeData['name'].values
labels
```

```python
# PCA
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
transData = pca.fit_transform(features, labels)
type(transData)
mergeData['pca1'] = DataFrame(transData[:,0])
mergeData['pca2'] = DataFrame(transData[:,1])
mergeData.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(x_vars=['pca1'], y_vars=['pca2'], data=mergeData, hue="class", size=5)
```

```python
textFrame = pd.DataFrame(features, columns = word)
textFrame.head()
len(features)
```

```python
allF = features.flatten()
nonZero = allF[np.nonzero(allF)]
m = np.median(nonZero)
```

```python
m = 0.1
dataset = list()
for i in range(len(features)):
    temp = textFrame.columns.values[[textFrame.iloc[i] >= m]]
    dataset.append(temp.tolist())
```

```python
for i in range(len(features)):
    print(dataset[i])
```

```python
# Apriori: https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
graph = df.T
graph.columns = mergeData['name']
```

```python
graph.head()
```

```python
from mlxtend.frequent_patterns import apriori

minsup = 0.01
apriori(graph, min_support = minsup, use_colnames=True)
```
